```python
def apply_prep(df):
    useless_cols = ['Patient_ID','Chip_Code','Chip_Image_Name','French_Residence_Department']
    cat_cols = ['Chip_Type', "French_Region", "Rural_or_urban_area"]
    obj_cols = ['Food_Type_0', 'Treatment_of_athsma', 'Age_of_onsets','General_cofactors','Treatment_of_atopic_dematitis']
    df = df.drop(useless_cols,axis=1)
    df = df.drop(obj_cols, axis=1)
    df = pd.get_dummies(df, columns = cat_cols)
    return df
```
